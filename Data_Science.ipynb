# 2.Mechine Learning

## 2.1 Supervised Learning 
### ▶2.1.1 Import Library & Decision Tree

```python
import matplotlib
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier 
select = ['Salary']  # ให้ทำการรับข้อมูล
target = ['Experience Level','Company Size','Employment Type']  # ให้ทำการส่งข้อมูล
x = df[select].head(70) # เลือกเเถวบน 70 เเถว
y = df[target].head(70)
dtree = DecisionTreeClassifier()
dtree = dtree.fit(x,y)
tree.plot_tree(dtree, feature_names=select)
plt.rcParams['figure.figsize'] = (120, 20) # ปรับขนาดของ decision tree
```

**นำ library matplotlib เเละ matplotlib.pyplot เข้ามาโดยให้ชื่อว่า plt เเล้วนำ library tree กับ DecisionTreeClassifier จาก sklearn , sklearn.tree จากนั้นให้เลือกคอลัมน์ที่ต้องการทำ decision tree โดยจะให้มีคอลัมน์ที่รับข้อมูลเเละส่งข้อมูล**

###▶2.1.2 Predict of Decision Tree

```python
print(dtree.predict([[100000]]))     # ใส่ข้อมูลของคอลัมน์ Salary 
print(dtree.predict([[200000]]))
```

**ให้ทำการใส่ข้อมูลของคอลัมน์ Salary เพื่อให้ทำนาย ข้อมูลของคอลัมน์ Experience Level , Company Size เเละ Employment Type ออกมา**

### ▶2.1.3  Linear Regression

```python
from scipy import stats
x = df['Work year']               # Work year
y = df['Salary in USD']          # Salary in USD

slope, intercept, r, p, std_err = stats.linregress(x, y) # ความชัน , จุดตัด , ค่าสัมประสิทธฺ์ ,ค่าสำหรับการทดสอบสมมติฐาน , ความผิดพลาดทางมาตรฐาน นำมาหาในกราฟ Linear Regression

def myfunc(x): # ฟังก์ชันหาค่าเชิงเส้น
  return slope * x + intercept

mymodel = list(map(myfunc, x)) # ให้วนซ้ำของค่าเชิงเส้น

plt.scatter(x, y)
plt.xlabel('Years')
plt.ylabel('Salary in USD')
plt.plot(x, mymodel)
plt.rcParams['figure.figsize'] = (5, 2)
plt.show()
```

**นำ library stats จาก scipy เเละทำการเลือกคอลัมน์ที่ต้องการ จากนั้นนำคอลัมน์มา Plot กราฟออกมาให้เป็น Linear Regression**

###▶2.1.4  Find Linear value

```python
year = myfunc(2020) # หาค่าเชิงเส้น
print(year)
year = myfunc(2021)
print(year)
year = myfunc(2022)
print(year)
year = myfunc(2023)
print(year)
```

**ทำการหาค่าเชิงเส้นเงินเดือนของเเต่ละปี**

# **7.รายได้สกุลเงิน USD จากอาชีพ Data Science ในเเต่ละปีมีค่าเฉลี่ยเท่าใด**
### ▶3.9 ฺBar Chart of Mean Salary in USD between year

```python
work_2020 = df.loc[(df['Work year'] == 2020)] #เลือกคอลัมน์ 'Work year' โดยเอาเเ่ต่ข้อมูลที่มีื่อว่า '2020'
work_2021 = df.loc[(df['Work year'] == 2021)] #เลือกคอลัมน์ 'Work year' โดยเอาเเ่ต่ข้อมูลที่มีื่อว่า '2021'
work_2022 = df.loc[(df['Work year'] == 2022)] #เลือกคอลัมน์ 'Work year' โดยเอาเเ่ต่ข้อมูลที่มีื่อว่า '2022'
work_2023 = df.loc[(df['Work year'] == 2023)] #เลือกคอลัมน์ 'Work year' โดยเอาเเ่ต่ข้อมูลที่มีื่อว่า '2023'
year_salary = pd.DataFrame(columns = ['2020', '2021', '2022', '2023']) # สร้างข้อมูลให้มีชื่อว่า 2020 2021 2022 2023
year_salary['2020'] = work_2020.groupby('Work year').mean('Salary in USD')['Salary in USD'].values # จับรวมกลุ่มคอลัมน์ 'Work year' เเละนำข้อมูลคอลัมน์ 'Salary in USD' มาหาค่าเฉลี่ย 
year_salary['2021'] = work_2021.groupby('Work year').mean('Salary in USD')['Salary in USD'].values
year_salary['2022'] = work_2022.groupby('Work year').mean('Salary in USD')['Salary in USD'].values
year_salary['2023'] = work_2023.groupby('Work year').mean('Salary in USD')['Salary in USD'].values
fig = go.Figure(data=px.bar(x = year_salary.columns, 
                            y = year_salary.values.tolist()[0],
                            color = year_salary.columns,
                            title = 'Mean Salary by Work Year'))

fig.update_layout(xaxis_title = "Work Year", yaxis_title = "Mean Salary (k)")
fig.show()
```

**นำคอลัมน์ชอง Work year มาเลือกเฉพาะเเต่ทีละข้อมูล เเละสร้างข้อมูลให้มีชื่อข้อมูลตรงกับคอลัมน์ 'Work year' เเล้วทำการจับกลุ่มของ 'Work year' เเละนำข้อมูลคอลัมน์ 'Salary in USD' หาค่าเฉลี่ย สุดท้ายนำมาเก็บในเเต่ละตัวแปร เพื่อให้ทำการสร้างกราฟเเท่งออกมา  **
